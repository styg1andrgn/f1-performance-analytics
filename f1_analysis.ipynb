```python
import pyodbc
import pandas as pd
from sqlalchemy import create_engine

driver = "ODBC Driver 17 for SQL Server"  # adjust based on what pyodbc.drivers() showed
server = "localhost"  # adjust if you use a named instance like localhost\SQLEXPRESS
database = "F1Analytics"

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    f"?driver={driver.replace(' ', '+')}"
    f"&trusted_connection=yes"
)

engine = create_engine(connection_string)

# Test it
df_test = pd.read_sql("SELECT TOP 5 * FROM vw_RaceResults", engine)
print(df_test)
```

```python
race_results = pd.read_sql("SELECT * FROM vw_RaceResults", engine)
driver_standings_history = pd.read_sql("SELECT * FROM vw_DriverStandingsHistory", engine)

print(race_results.shape)
print(driver_standings_history.shape)
print(race_results.head())
```

```python
race_results = race_results.sort_values(['driver_id', 'season', 'round']).reset_index(drop=True)
```

```python
race_results['recent_form_points'] = (
    race_results.groupby('driver_id')['points']
    .transform(lambda x: x.shift(1).rolling(window=3, min_periods=1).mean())
)
```

```python
race_results = race_results.sort_values(['constructor_id', 'season', 'round']).reset_index(drop=True)

race_results['constructor_recent_form'] = (
    race_results.groupby('constructor_id')['points']
    .transform(lambda x: x.shift(1).rolling(window=3, min_periods=1).mean())
)
```

```python
print(race_results['recent_form_points'].isna().sum())
print(race_results['constructor_recent_form'].isna().sum())
```

```python
race_results['recent_form_points'] = race_results['recent_form_points'].fillna(0)
race_results['constructor_recent_form'] = race_results['constructor_recent_form'].fillna(0)

# Confirm no NaNs remain
print(race_results['recent_form_points'].isna().sum())
print(race_results['constructor_recent_form'].isna().sum())
```

```python
# Filter to current points system era
race_results = race_results[race_results['season'] >= 2010].copy()

# Recompute driver recent form
race_results = race_results.sort_values(['driver_id', 'season', 'round']).reset_index(drop=True)
race_results['recent_form_points'] = (
    race_results.groupby('driver_id')['points']
    .transform(lambda x: x.shift(1).rolling(window=3, min_periods=1).mean())
)

# Recompute constructor recent form
race_results = race_results.sort_values(['constructor_id', 'season', 'round']).reset_index(drop=True)
race_results['constructor_recent_form'] = (
    race_results.groupby('constructor_id')['points']
    .transform(lambda x: x.shift(1).rolling(window=3, min_periods=1).mean())
)

# Fill debut NaNs
race_results['recent_form_points'] = race_results['recent_form_points'].fillna(0)
race_results['constructor_recent_form'] = race_results['constructor_recent_form'].fillna(0)

print(race_results.shape)
print(race_results['season'].min(), race_results['season'].max())
```

```python
train = race_results[race_results['season'] <= 2022]
test = race_results[race_results['season'] > 2022]

print("Train shape:", train.shape)
print("Test shape:", test.shape)
print("Train seasons:", sorted(train['season'].unique()))
print("Test seasons:", sorted(test['season'].unique()))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

features = ['grid', 'recent_form_points', 'constructor_recent_form']
target = 'points'

X_train, y_train = train[features], train[target]
X_test, y_test = test[features], test[target]

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

print("MAE:", mean_absolute_error(y_test, predictions))
print("R²:", r2_score(y_test, predictions))

# Also useful: look at the model's coefficients
for feature, coef in zip(features, model.coef_):
    print(f"{feature}: {coef:.3f}")
print("Intercept:", model.intercept_)
```

```python
print(y_test.describe())
```

```python
test = test.copy()
test['predicted_points'] = predictions
test['residual'] = test['points'] - test['predicted_points']

print(test['residual'].describe())
```

```python
residual_std = test['residual'].std()
threshold = 2 * residual_std

test['anomaly'] = test['residual'].abs() > threshold
print(f"Residual std: {residual_std:.2f}, threshold: ±{threshold:.2f}")
print(f"Number of anomalies flagged: {test['anomaly'].sum()} out of {len(test)}")
```

```python
anomaly_cols = ['race_name', 'season', 'round', 'driver_name', 'constructor_name', 
                 'grid', 'position_order', 'points', 'predicted_points', 'residual', 'status']

anomalies = test[test['anomaly']].sort_values('residual')
print(anomalies[anomaly_cols].to_string(index=False))
```

```python
race_results['predicted_points'] = model.predict(race_results[features])
race_results['residual'] = race_results['points'] - race_results['predicted_points']
race_results['data_split'] = race_results['season'].apply(lambda s: 'train' if s <= 2022 else 'test')

# Only compute anomaly threshold/flags using out-of-sample (test) residuals
test_residual_std = race_results.loc[race_results['data_split'] == 'test', 'residual'].std()
threshold = 2 * test_residual_std

race_results['anomaly'] = (
    (race_results['data_split'] == 'test') & 
    (race_results['residual'].abs() > threshold)
)

print(f"Threshold (based on test residuals): ±{threshold:.2f}")
print(race_results['anomaly'].value_counts())
```

```python
race_results.to_sql('model_results', engine, if_exists='replace', index=False)
```
